```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
city = input('enter the city')
cityArea = input('enter the area')
```

```python
def get_data():
    df = pd.DataFrame(columns=['name','link'])
    url = 'https://housing.com/in/buy/'+ str(city) + '/' + str(cityArea)
    response = requests.get(url)
    soup=BeautifulSoup(response.content,'html.parser')
    for d in soup.find_all('div', attrs={'class':'T_cardV1Style _bx1t02 _gwftgi'}):
        data = []
        names = d.find('div', attrs ={'class':'title-style'})
        # inputs = soup.find_all('input', {'type': 'button'})
        # button = soup.find('button', {'data-testid': 'buttonId'}) 
        # print("Text:", button.text)
        # print("Attributes:", button.attrs)
        if names is not None:
            data.append(names.text)
            print(names.text)
        else:
           data.append('NAN')
        link = d.find('a')
        if link is not None:
            data.append('https://housing.com/'+ link.get('href'))
            print('https://housing.com/'+ link.get('href'))
        else:
            data['link'] = 'NAN'
        df.loc[len(df)] = data
    return df
```

```python
get_data()
```

```python
df.to_csv('housing.csv', index=False)
```
